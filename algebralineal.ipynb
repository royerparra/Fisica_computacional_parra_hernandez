__Ejercicio 2:__ ¿Como se define una matriz simétrica?

__Solución:__ definimos una matriz $A\in \mathbb{R}^{nxn}$ a estas las llamamos cuadradas, por lo cual para este tipo especial denotamos sus entradas como $A=a_{ij}$ donde $i$ denota el número de renglón y $j$ el de columna en la matriz $A$, definimos la matriz $A^T=a_{ji}$ ahora definimos que $A$ es simétrica cuando ocurre que $A=A^T$, note como los renglones de $A$ son las columnas de $A^T$ y veceversa.

__Ejercicio 3:__ ¿Como puede construir una matriz simétrica a partir de cualquier matriz cuadrada?

__Solución:__ no es muy difícil construir una matriz simétrica a partir de una matriz cuadrada cualquiera, sin embargo para ver que nuestra definición que vamos a dar es consistente necesitamos la siguiente observación:

Sea $A,B\in \mathbb{R}^{nxn}$ entonces el producto de matrices $AB$ es $ab_{ij}=\sum_{k=1}^{n}a_{ik}b_{kj}\equiv a_{ik}b_{kj}$, donde en la última ecuación se emplea la convención de einstein que indica suma sobre índices repetidos, es fácil observar que la entrada $ij$ del producto de las matrices es el producto punto entre el renglón $i$ de la primer matriz denotado por $A_i$ y la columna $j$ de la segunda matriz denotado por $B^j$, de manera que se puede reescribir como $A_i\cdot B^j$, además por la observación del inciso anterior $A^j=A^T_j$ y $A_i=(A^T)^i$

por último por razones de consistencia necesitamos los siguientes resultados:

Sea $A\in\mathbb{R}^{nxn}$ entonces ocurre que: $i)$ $\left(A^T\right)^T=A$ $ii)$ $(AB)^T=B^TA^T$:

Demostración: i) $\left(A^T\right)^T=((a_{ij})^T)^T=(a_{ji})^T=a_{ij}=A$

ii) se tiene que $(AB)^T=(a_{ik}b_{kj})^T=a_{jk}b_{ki}=b_{ki}a_{jk}=A_j\cdot B^i=(A^T)^j\cdot (B^T)_i=(B^T)_i\cdot (A^T)^j=B^TA^T$ esto último se sigue de las observaciones del parrafo anterior.

ahora estamos equipados suficientemente para escoger la siguiente definición:

Sea $A\in\mathbb{R}^{nxn}$ entonces la matriz asociada a $A$ denotada por $B=AA^T$ es una matriz simétrica, esto se sigue de que $B^T=(AA^T)^T=(A^T)^TA^T=AA^T=B$

__Ejercicio 4:__ Construya el propagador $e^{A t}$ para cada tiempo sobre una malla y uselo para propagar un estado inicial $\vec x_0$ cualquiera. Utilice:
$$A=\left(\begin{array}{cc} -1 & 1\\ -1 & -1 \end{array} \right).$$
* Grafique la trayectoria en el espacio de fase.
* Recuerde que $\vec x(t) = e^{A t}\vec x_0$, y necesitamos graficar la trayectoria $\vec x(t)$ en el espacio de fase.

__Nota 1:__ De momento no utilicen `expm`.

__Nota 2:__ El operador `*` funciona para multiplicar matrices.

__Note 3:__ Observen que la matriz $A$ no es simétrica, aqui no se cumple que $w^T=w^{-1}$.

__solución:__ en primer lugar vamos a ocupar el resultado probado en la tarea pasada, el cual nos dice que si $B=WAW^{-1}$ matrices entonces para cualquier función analítica $f(B)=Wf(A)W^{-1}$ de manera que si nuestra $A$ fuera diagonal nos simplifica muchísimo el trabajo ya que sabemos que $f(diag(a_1,\ldots,a_n))=diag(f(a_1),\ldots,f(a_n))$, por tanto vamos a diagonalizar la matriz A.

```julia
A=[-1 1; -1 -1]
```

```julia
eigvals(A)# calculamos usando julia los eigenvalores de A
```

comprobamos que en efecto estos son los eigenvalores pues $\lambda_{1,2}=-1\pm i$, de manera que :

$\left|\begin{array}{cc} -1+1\pm i & 1\\ -1 & -1+1\pm i \end{array} \right|=\left|\begin{array}{cc} \pm i & 1\\ -1 & \pm i \end{array} \right|=(\pm i)(\pm i)-(1)(-1)=i^2+1=0$
ahora procedemos a diagonalizar, para eso buscamos los eingenvalores asociados a $\lambda_1=1-i$ y $\lambda_2=1+i$:

$\lambda_1)$ en este caso tenemos que:

$\left(\begin{array}{cc} i & 1\\ -1 & i \end{array} \right)  \left(\begin{array}{cc} x\\ y \end{array} \right)=\left(\begin{array}{cc} 0 \\ 0 \end{array} \right) \Rightarrow \left(\begin{array}{cc} ix+y \\ -x+iy \end{array} \right)=\left(\begin{array}{cc} 0 \\ 0 \end{array} \right)$

de esta manera obtenemos que el eigenvector asociado es de la forma  $(iy,y)=y(i,1)$ ahora pedimos que su norma sea $1$, es decir que $|y||(i,1)|=1$,de aqui se obtiene que $|y|=\frac{1}{\sqrt 2}$, de donde nuestro vector que esocgemos es:

$$v_1=\frac{1}{\sqrt 2} \left(\begin{array}{cc} i \\ 1 \end{array} \right)$$

$\lambda_2)$ ahora se tiene que:

$\left(\begin{array}{cc} -i & 1\\ -1 & -i \end{array} \right)  \left(\begin{array}{cc} x\\ y \end{array} \right)=\left(\begin{array}{cc} 0 \\ 0 \end{array} \right) \Rightarrow \left(\begin{array}{cc} -ix+y \\ -x-iy \end{array} \right)=\left(\begin{array}{cc} 0 \\ 0 \end{array} \right)$

de manera analoga y omitiendo pasos obtenemos que en este caso escogemos:

$$v_2=\frac{1}{\sqrt 2} \left(\begin{array}{cc} -i \\ 1 \end{array} \right)$$

ahora vamos a observar lo siguiente con diagramas:

usando lenguaje de álgebra lineal, $A$ es la representación canónica de un operador $T$, es decir, es la reprensentación en la ya conocida base $\{\hat{e_1},\hat{e_2}\}$ a la cual llamamos sistema $S$, pero ya sabemos que nuestro operador $T$ tiene una representación diagonal en la base $\{v_1,v_2\}$ a este lo llamamos $S'$ de manera que las matrices que diagonalizan $A$ son las del cambio de coordenadas (ella y su inversa), por tanto tenemos el siguiente diagrama para la diagonalización $S\rightarrow S'$ (cambio de coordenadas) $\rightarrow S'$ (se aplica el operador T que ahora es diagonal)$\rightarrow S$ (cambio de coordenadas de "regreso") 

observemos que $\langle v_1,v_2\rangle=1/2((-i)*(-i)+1*1)=1/2(-1+1)=0$ es decir son ortogonales ( recuerde que $\langle,\rangle$ denota al producto hermitiano estandar en $\mathbb{C}$), pero además es fácil ver que $\langle v_1,v_1\rangle=\langle v_2,v_2\rangle=1$ de manera que forman una base ortonormal en $\mathbb{C}^2$, por tanto por conocimientos de Lineal II sabemos que la matriz de cambios de coordenadas en unitaria, es decir $W^*=W^{-1}$ donde $w^*=\tilde{w^T}$ donde la tilde senaña conjugación, por último es facil ver que la matriz que realiza el cambio de coordenadas de $S'$ a $S$ es:
$$W=\frac{1}{\sqrt 2} \left(\begin{array}{cc} i & -i\\ 1 & 1 \end{array} \right)$$
de manera que tenemos por la observación de álgebra Lineal II:
$$W^{-1}=\frac{1}{\sqrt 2} \left(\begin{array}{cc} -i & 1 \\ i & 1 \end{array} \right)$$
 así pues se tiene que definimos:
$$D\equiv \left(\begin{array}{cc} 1-i & 0 \\ 0 & 1+i \end{array} \right)$$
de esta manera usando el diagrama de mas arriba obtenemos que $A=WDW^{-1}$ así pues $e^{At}=We^{Dt}W^{-1}$, ahora procedemos a hacer el cálculo:
 
$$e^{Dt}=\left(\begin{array}{cc} e^{at} & 0 \\ 0 & e^{\bar{a}t} \end{array} \right)$$
donde $a=1-i$, de esta manera tenemos que:
$$e^{At}=\frac{1}{\sqrt 2} \left(\begin{array}{cc} i & -i \\ 1 & 1 \end{array} \right)\left(\begin{array}{cc} e^{at} & 0 \\ 0 & e^{\bar{a}t} \end{array} \right) \frac{1}{\sqrt 2} \left(\begin{array}{cc} -i & 1 \\ i & 1 \end{array} \right)=\frac{1}{2}\left(\begin{array}{cc} i & -i \\ 1 & 1 \end{array} \right)\left(\begin{array}{cc} -ie^{at} & e^{at} \\ ie^{\bar{a}t} & e^{\bar{a}t} \end{array} \right)=\frac{1}{2}\left(\begin{array}{cc} e^{at}+e^{\bar{a}t} & i(e^{at}-e^{\bar{a}t}) \\ -i(e^{at}+e^{\bar{a}t}) & e^{at}+e^{\bar{a}t} \end{array} \right)$$
pero ahora recordemos que $e^{at}\pm e^{\bar{a}t}=e^{-t-it}\pm e^{-t+it}=e^{-t}(\cos(t)-i\sin(t))\pm e^{-t}(\cos(t)+i\sin(t))=e^{-t}(\cos(t)-i\sin(t)\pm \cos(t)+i\sin(t))$ de manera que $e^{at}+e^{\bar{a}t}=2e^{-t}\cos(t)$ y $e^{at}-e^{\bar{a}t}=-2ie^{-t}\sin(t)$, por tanto tenemos que:
$$e^{At}=e^{-t}\left(\begin{array}{cc} \cos(t) & \sin(t) \\ -\sin(t) & \cos(t) \end{array} \right)$$
de manera que dado un estado estacionario $\vec x_0=(x',y')$, lo podemos propagar en el tiempo así:
$$\vec{x_0}(t)=e^{-t}(x'\cos(t)+y'\sin(t),-x'\sin(t)+y'cos(t))$$
una observación que se puedo notar en seguida es que el propagador en este caso es una rotación que depende del tiempo, pero además se tiene un factor reductivo $e^{-t}$ esto pues se considera $t\geq0$ de manera que siempre es menor o igual que $1$, de esta manera lo que uno espera es que de por resultado espirales que colapsan hacía el centro, es decir, uno espera que el $\vec0$ sea un atractor global.
Ahora procedemos a construir nuestro propagador con codigo y así graficar el espacio fase

```julia
G(t)=[exp(-t)*cos(t) exp(-t)*sin(t); -exp(-t)*sin(t) exp(-t)*cos(t)]#este es nuestro propagador en el tiempo
```

```julia
G(pi)#en verdad se puede aplicar en algún tiempo dado
```

```julia
a=[1,1] #vamos a hacer una pruebas y observaciones
```

```julia
b=G(pi)*a #si lo aplicamos a este tipo de vector, no hay ningún problema
```

```julia
b[2,1] #esto nos da el segundo renglón del vector b
```

```julia
b[1,1] #esto nos da el primer renglón del vector b
```

ahora vamos a graficar en el espacio de fases tomando en cuenta valores de $\vec x_0$ aleatorios

```julia
using PyPlot
```

```julia
t=linspace(0,2*pi,10000)#usaremos este linspace para graficar usando el intervalo [0,2*pi], o bien dicho de otro modo usando la observación anterior, graficamos en una sola vuelta
```

```julia
title("espacio de fases")
for i in 1:1000#graficaremos 1000 condiciones aleatorias
    M=2*rand(2,1)-[1,1]#creamos vectores aleatorios en el rectangulo [-1,1]X[-1,1]
    X=[]#estas serán nuestras listas en x e y de la trayectorio M(t)
    Y=[]
    for j in t#para cada valor del linspace 
        R=G(j)*M#calculamos el vector propagado
        push!(X,R[1,1])#guardamos en X su coordenada x
        push!(Y,R[2,1])#y en Y su coordenada y
    end
    plot(X,Y)#graficamos ambas listas
end
```

Observe como en efecto se cumplió la predicción hecha al inicio con solo mirar la matriz, es decir, son espirales que convergen al centro $(0,0)$ al irse a infinito, es decir, el origen es un atractor global, pues como se puede ver de la expresión del propagador, es evidente que sin importar la condición inicial cuando $t\rightarrow\infty$, $\vec{x}(t)\rightarrow \vec0$.
