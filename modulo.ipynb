```julia
h(x,t)=0#probamos para x'=0 es decir las constantes
```

```julia
l=linspace(0,1,10000)
```

```julia
a=herramientas.RK4(h,l,1)
```

```julia
using PyPlot
```

```julia
title("prueba")
plot(l,a)
```

```julia
p(x,t)=x#ahora probamos para la ecuacion x'=x
```

```julia
title("soluciones exponenciales")
for i in 1:1000#aplicamos 1000 condiciones inciales
    w=rand()
    b=herramientas.RK4(p,l,w)
    plot(l,b)
end
```

```julia
q(x,t)=4-cos(x)#por último probamos esta.
```

```julia
title("soluciones")
for i in 1:2000#aplicamos 2000 condiciones inciales
    w=2*rand()-1
    b=herramientas.RK4(q,l,w)
    plot(l,b)
end
```
